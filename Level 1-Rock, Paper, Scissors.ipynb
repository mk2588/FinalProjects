Create a program where the user can play a game of rock, paper, scissors against a computer. Set a play choice (rock, paper, or scissors) for the computer player and during the program, request for the user to input their play choice.

The rules in Rock, Paper, Scissors are:

Rock beats scissors
Scissors beats paper
Paper beats rock
If the user wins, then display "You win!" If they lose against the computer, then display "You lose." If the computer and the user have the same play choice, display "It's a draw."

Also ask the user if they would like to play again. If they do, start the game again but if not, thank the user for playing the game.

```python
print("Rock, Papers, Scissors!")

import random as rand

while True:
    actions = ["rock","paper","scissors"]
    play_action = input("Please enter your choice (rock, paper, scissors): ") # player's choice
    comp_choice = rand.randint(0,2) # computer's choice
    comp_action = actions[comp_choice]
    print(f"Computer's choice: {actions[comp_choice]}")

# Possible results
    if play_action == comp_action:
        print("It's a draw.")
    elif play_action == "rock":
        if comp_action == "scissors":
            print("You win!")
        else:
            print("You lose.")
    elif play_action == "paper":
        if comp_action == "scissors":
            print("You lose.")
        else:
            print("You win!")
    elif play_action == "scissors":
        if comp_action == "rock":
            print("You lose.")
        else:
            print("You win!")
    else:
        print("This is not a valid choice, please choose again.")

    again = input("Would you like to play again? (y/n) ")
    if again.lower() == "n":
        print("Thank you for playing the game.")
        break
```
